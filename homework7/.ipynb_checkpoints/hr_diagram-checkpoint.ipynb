```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.loadtxt("star_data.txt", skiprows=1) # loads data in numpy array
```

```python
#3.2 
data.shape #2642 rows
```

```python
magnitude = data[:, 0]
temperature = data[:, 1]
```

```python
minimmag = magnitude.min()
maximmag = magnitude.max()
print(minimmag)
print(maximmag)
minimagindex = np.argmin(magnitude)
print(minimagindex)
maxmagindex = np.argmax(magnitude)
print(maxmagindex)
# how do i know what stars they correspond to??
```

```python
minitemp = temperature.min()
maxitemp = temperature.max()
print(minitemp)
print(maxitemp)
minitempindex = np.argmin(temperature)
maxmagindex = np.argmax(temperature)
print(minitempindex)
print(maxmagindex)
```

```python
# star with 2.24 mag/number 1988 is brightest, 12.49/number 264 is dimmest
log_magnitude = np.log(magnitude)
log_temperature = np.log(temperature)
plt.figure(figsize=(8,6))
plt.scatter(log_magnitude, log_temperature)
plt.gca().invert_xaxis()
plt.gca().invert_yaxis()
brightest = np.argmin(magnitude)
dimmest = np.argmax(magnitude)
hottest = np.argmax(temperature)
coldest = np.argmin(temperature)
plt.scatter(log_magnitude[brightest], log_temperature[brightest], color='black', marker='*', label='Brightest')
plt.scatter(log_magnitude[dimmest], log_temperature[dimmest], color='purple', marker='*', label='Dimmest')
plt.scatter(log_magnitude[hottest], log_temperature[hottest], color='green', marker='*', label='Hottest')
plt.scatter(log_magnitude[coldest], log_temperature[coldest], color='orange', marker='*', label='Coldest')
plt.xlabel('Temperature(log K)')
plt.ylabel('Magnitude')
plt.title('Hertzsprug-Russel Diagram')
plt.legend()
```
