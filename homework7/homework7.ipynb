```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# 2 Matplotlib
def make_sine_wave(x, A, w):
    return A * np.sin(x*w)
```

```python
x = np.linspace(0, 2*np.pi, 1000)
amplitude = [0.5, 1, 1.5, 2, 2.5]
frequency = [1, 2, 3, 4, 5]
fig, ax = plt.subplots(1, 3, figsize = (25,5))
```

```python
tarray = zip(amplitude, frequency)
for A, w in tarray:
    var = make_sine_wave(x, A, 1)
    ax[0].plot(x, var)
ax[0].set_xlabel('x')
ax[0].set_ylabel('y')
for A, w in zip(amplitude, frequency):
    var = make_sine_wave(x, 1, w)
    ax[1].plot(x, var)
ax[1].set_xlabel('x')
ax[1].set_ylabel('y')
for A, w in zip(amplitude, frequency):
    var = make_sine_wave(x, A, w)
    ax[2].plot(x, var)
ax[2].set_xlabel('x')
ax[2].set_ylabel('y')
# ax[0].legend()
# ax[1].legend()
# ax[2].legend()
# plt.xlabel('x')
plt.show()
```
